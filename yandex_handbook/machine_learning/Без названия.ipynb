**Метрика accuracy (Точность)** -  это доля верных предсказаний среди всех предсказаний. Она показывает, насколько хорошо модель классифицировала все примеры, независимо от их классов.

$$ \text{Accuracy} = \frac{\text{Число верных предсказаний}}{\text{Общее число предсказаний}} $$
Когда использовать:

Accuracy подходит, если классы сбалансированы (например, 50% положительных и 50% отрицательных примеров), и если одинаково важно правильно классифицировать оба класса.<br>
Пример:

Если у нас есть 100 примеров, и модель правильно классифицировала 90 из них, то:

$$ \text{Accuracy} = \frac{90}{100} = 0.9 \; \text{или 90%} $$



**Метрика Precision (Точность предсказания)** — это доля верных положительных предсказаний среди всех предсказанных положительных. Она показывает, насколько модель уверена в своих положительных предсказаниях.

$$ \text{Precision} = \frac{\text{Число истинно положительных предсказаний}}{\text{Число всех положительных предсказаний}} $$

Когда использовать:

Precision важна, если ложноположительные ошибки нежелательны. Например, в задачах, где ошибочно положительный результат может быть более опасен, чем ошибочно отрицательный (например, при диагностике редких заболеваний или в задачах, где модель должна определить спам).

Пример:

Если модель предсказала 20 положительных случаев, но только 15 из них были действительно положительными, то:

$$\text{Precision} = \frac{15}{20} = 0.75 \; \text{или 75%}$$



**Метрика Recall (Полнота, чувствительность)** — это доля истинно положительных предсказаний среди всех реальных положительных примеров. Она показывает, насколько хорошо модель находит все положительные примеры.


$$ \text{Recall} = \frac{\text{Число истинно положительных предсказаний}}{\text{Число всех реальных положительных примеров}} $$


Когда использовать:

Recall важен, если пропущенные положительные примеры нежелательны. Это важно, например, в задачах диагностики болезней, где нужно определить всех больных пациентов (даже если это означает больше ложноположительных).

Пример:

Если в выборке 30 положительных случаев, и модель нашла 15 из них, то:

$$ \text{Recall} = \frac{15}{30} = 0.5 \; \text{или 50%} $$
